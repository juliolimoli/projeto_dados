# Preparacao das tabelas treino e teste

```python
import pandas as pd
import os
```

## Importando a tabela filtrada no databricks

```python
df = pd.read_csv('./bases/application_train_filtered.csv')
df.head()
```

## Separando em treino e teste

```python
def separar_treino(x, df_treino):
    if x.name in df_treino.index:
        return True
    else:
        return False

df_treino = df.sample(frac=0.7, random_state=42)
df['Flag_Treino'] = df.apply(lambda x: separar_treino(x, df_treino), axis=1)
df['Flag_Treino'].value_counts()
```

## Salvando os datasets

```python
os.makedirs('./bases', exist_ok=True)  

df.to_csv('./bases/application_splitado_07.csv', index=False)
```
